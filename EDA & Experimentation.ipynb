```python
import time
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error

from sklearn.model_selection import GridSearchCV



param_grid_decision_tree = {
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

param_grid_random_forest = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

param_grid_xgboost = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'n_estimators': [50, 100, 200]
}

param_grid_lightgbm = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'n_estimators': [50, 100, 200]
}

param_grid_catboost = {
    'learning_rate': [0.01, 0.1, 0.2],
    'depth': [3, 5, 7],
    'n_estimators': [50, 100, 200]
}

param_grid_mlp_regressor = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 50)],
    'activation': ['relu', 'tanh'],
    'learning_rate_init': [0.001, 0.01, 0.1]
}

# Create GridSearchCV objects for each model
grid_search_decision_tree = GridSearchCV(DecisionTreeRegressor(), param_grid_decision_tree, scoring='neg_mean_squared_error', cv=5)
grid_search_random_forest = GridSearchCV(RandomForestRegressor(), param_grid_random_forest, scoring='neg_mean_squared_error', cv=5)
grid_search_xgboost = GridSearchCV(XGBRegressor(), param_grid_xgboost, scoring='neg_mean_squared_error', cv=5)
grid_search_lightgbm = GridSearchCV(LGBMRegressor(), param_grid_lightgbm, scoring='neg_mean_squared_error', cv=5)
grid_search_catboost = GridSearchCV(CatBoostRegressor(), param_grid_catboost, scoring='neg_mean_squared_error', cv=5)
grid_search_mlp_regressor = GridSearchCV(MLPRegressor(), param_grid_mlp_regressor, scoring='neg_mean_squared_error', cv=5)

# Measure the time for training each model
start_time = time.time()
grid_search_decision_tree.fit(x_train, y_train)
end_time = time.time()
print(f"Decision Tree Training Time: {end_time - start_time} seconds")

start_time = time.time()
grid_search_random_forest.fit(x_train, y_train)
end_time = time.time()
print(f"Random Forest Training Time: {end_time - start_time} seconds")

start_time = time.time()
grid_search_xgboost.fit(x_train, y_train)
end_time = time.time()
print(f"XGBoost Training Time: {end_time - start_time} seconds")

start_time = time.time()
grid_search_catboost.fit(x_train, y_train)
end_time = time.time()
print(f"CatBoost Training Time: {end_time - start_time} seconds")

start_time = time.time()
grid_search_mlp_regressor.fit(x_train, y_train)
end_time = time.time()
print(f"MLP Regressor Training Time: {end_time - start_time} seconds")

# Print best parameters and corresponding MSE for each model
print("Decision Tree Best Parameters:", grid_search_decision_tree.best_params_)
print("Decision Tree Best MSE:", -grid_search_decision_tree.best_score_)

print("Random Forest Best Parameters:", grid_search_random_forest.best_params_)
print("Random Forest Best MSE:", -grid_search_random_forest.best_score_)

print("XGBoost Best Parameters:", grid_search_xgboost.best_params_)
print("XGBoost Best MSE:", -grid_search_xgboost.best_score_)

print("CatBoost Best Parameters:", grid_search_catboost.best_params_)
print("CatBoost Best MSE:", -grid_search_catboost.best_score_)

print("MLP Regressor Best Parameters:", grid_search_mlp_regressor.best_params_)
print("MLP Regressor Best MSE:", -grid_search_mlp_regressor.best_score_)
```

```python
import time
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error

from sklearn.model_selection import GridSearchCV

import warnings
warnings.filterwarnings("ignore")

param_grid_decision_tree = {
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

param_grid_xgboost = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'n_estimators': [50, 100, 200]
}

param_grid_lightgbm = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'n_estimators': [50, 100, 200]
}

param_grid_catboost = {
    'learning_rate': [0.01, 0.1, 0.2],
    'depth': [3, 5, 7],
    'n_estimators': [50, 100, 200]
}

param_grid_mlp_regressor = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 50)],
    'activation': ['relu', 'tanh'],
    'learning_rate_init': [0.001, 0.01, 0.1],
    'max_iter': [200, 500, 1000]  # Increase the max_iter values
}


# Create GridSearchCV objects for each model
grid_search_decision_tree = GridSearchCV(DecisionTreeRegressor(), param_grid_decision_tree, scoring='neg_mean_squared_error', cv=5)
grid_search_xgboost = GridSearchCV(XGBRegressor(), param_grid_xgboost, scoring='neg_mean_squared_error', cv=5)
grid_search_lightgbm = GridSearchCV(LGBMRegressor(), param_grid_lightgbm, scoring='neg_mean_squared_error', cv=5)
grid_search_catboost = GridSearchCV(CatBoostRegressor(), param_grid_catboost, scoring='neg_mean_squared_error', cv=5)
grid_search_mlp_regressor = GridSearchCV(MLPRegressor(), param_grid_mlp_regressor, scoring='neg_mean_squared_error', cv=5)

# Measure the time for training each model
start_time = time.time()
grid_search_decision_tree.fit(x_train, y_train)
end_time = time.time()
print(f"Decision Tree Training Time: {end_time - start_time} seconds")

start_time = time.time()
grid_search_xgboost.fit(x_train, y_train)
end_time = time.time()
print(f"XGBoost Training Time: {end_time - start_time} seconds")

start_time = time.time()
grid_search_lightgbm.fit(x_train, y_train)
end_time = time.time()
print(f"Lightgbm Training Time: {end_time - start_time} seconds")

start_time = time.time()
grid_search_catboost.fit(x_train, y_train)
end_time = time.time()
print(f"CatBoost Training Time: {end_time - start_time} seconds")

start_time = time.time()
grid_search_mlp_regressor.fit(x_train, y_train)
end_time = time.time()
print(f"MLP Regressor Training Time: {end_time - start_time} seconds")

# Print best parameters and corresponding MSE for each model
print("Decision Tree Best Parameters:", grid_search_decision_tree.best_params_)
print("Decision Tree Best MSE:", -grid_search_decision_tree.best_score_)

print("XGBoost Best Parameters:", grid_search_xgboost.best_params_)
print("XGBoost Best MSE:", -grid_search_xgboost.best_score_)

print("LightGBM Best Parameters:", grid_search_lightgbm.best_params_)
print("LightGBM Best MSE:", -grid_search_lightgbm.best_score_)

print("CatBoost Best Parameters:", grid_search_catboost.best_params_)
print("CatBoost Best MSE:", -grid_search_catboost.best_score_)

print("MLP Regressor Best Parameters:", grid_search_mlp_regressor.best_params_)
print("MLP Regressor Best MSE:", -grid_search_mlp_regressor.best_score_)
```
